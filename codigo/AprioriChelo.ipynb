```python
import pandas as pd

df = pd.read_csv('BD/OnlineRetail_Preprocesada2.csv', sep=',', encoding = 'unicode_escape')

df.head()
```

```python
print(len(df))
```

```python
# este paso lo agrego porque el dataset preproceados tiene una primera columna que es considerada en el ejercicio
df.drop("Unnamed: 0", inplace=True, axis= 1)

df.head()
```

```python
print(len(df))
```

```python
# Cada una de las filas representa una factura y cada una de las columnas un producto. Esto se consigue utilizando el siguiente c√≥digo
basket = (df[df['Country']=="Spain"]
          .groupby(['InvoiceNo', 'Description'])['Quantity']
          .sum().unstack().reset_index().fillna(0)
          .set_index('InvoiceNo'))

basket.drop('POSTAGE', inplace=True, axis=1)
```

```python
basket.head()
```

```python
print(len(basket))
```

```python
#Se utiliza funci√≥n `apriori` para obtener los productos con un m√≠nimo de soporte.

from mlxtend.frequent_patterns import apriori

frequent_itemsets = apriori(basket > 0, min_support=0.06, use_colnames=True)
frequent_itemsets.head()
```

```python
print(len(frequent_itemsets))
```

```python
# se pueden obtener las reglas que superen adem√°s un m√≠nimo de confianza
from mlxtend.frequent_patterns import association_rules

reglas_minimo_confianza = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.8)
reglas_minimo_confianza.head()
```

```python
print(len(reglas_minimo_confianza))
```

```python
# tambi√©n desde el dataset generado con apriori se puede obtener reglas con un m√≠nimo de soporte

reglas_minimo_soporte = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
reglas_minimo_confianza.head()
```

```python
print(len(reglas_minimo_confianza))
```
