```python
#Se genera la matriz igual que en el "fase2" del repo 
#Recordar cambiar la direccion desde donde se está leyendo la csv ya que creé uan copia

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from apyori import apriori

df_bd_preprocesada = pd.read_csv('BD/OnlineRetail_Preprocesada2.csv', sep=',', encoding = 'unicode_escape')
df_pivot = df_bd_preprocesada['InvoiceDate'].str.split(expand = True)
df_pivot.rename(columns = {0: 'Fecha', 1: 'Hora'}, inplace=True)
df_bd_preprocesada = df_bd_preprocesada.drop(columns =['Unnamed: 0','InvoiceDate'])
df_bd_preprocesada = pd.concat([df_bd_preprocesada,df_pivot], axis = 1)
df_bd_preprocesada['Monto'] = df_bd_preprocesada['UnitPrice'] * df_bd_preprocesada['Quantity']
df_bd_ordenadas = df_bd_preprocesada.sort_values('InvoiceNo')
```

```python
#Se hace un head para ver como queda ordenado por "InoviceNo" de menor a mayor
df_bd_ordenadas.head(392734)
```

```python
#Largo de la tabla anterior
len_records = len(df_bd_ordenadas)
print(len_records)
```

```python
#Se intenta dejar la matriz igual a los que existen dentro de los ejemplos de apriori en internet

matriz = []
# antes era 22-532-64
maxF = 5
maxL = 532
maxC  = 20
for i in range(maxF):
    a = ["nan"]*maxC
    matriz.append(a)
invoice = df_bd_ordenadas.values[0,0]
ii = 0
jj = 0
for i in range (0,maxL):
    if invoice == df_bd_ordenadas.values[i,0] and jj < maxC:
        desc = str(df_bd_ordenadas.values[i,1])
        matriz[ii][jj]=desc
        jj += 1
    else:
        ii += 1
        if ii > maxF-1:
            break
        else:
            jj = 1
            desc = str(df_bd_ordenadas.values[i,1])
            matriz[ii][0]=desc
            invoice = df_bd_ordenadas.values[i,0]
print (matriz)    
```

```python
#La matriz resultante se deja en un nuevo CSV nuevo para despues pasarla como parametro

import csv
myFile = open('nueva_matriz.csv', 'w')
with myFile:
    writer = csv.writer(myFile)
    writer.writerows(matriz)
     
print("Escritura completa")
```

```python
#Se genera el nuevo CSV con un head para ver si el orden queda igual a los ejemplos

df_x = pd.read_csv('./nueva_matriz.csv', sep=',', encoding = 'unicode_escape', header = None)
df_x.head(392734)
```

```python
#Se imprime el largo de columnas
nums_reg = len(df_x)
print (nums_reg)
```

```python
#Se genera una matriz que es la que se pasará como parametro para la funcion apriori

records = []
for i in range(0, nums_reg):
    records.append([str(df_x.values[i,j]) for j in range(0, maxC-1)])
```

```python
#Se generan variables de tiempo para determinar cuando inicia y se termina tanto las reglas de asociacion como los resultados
#Se funciona a la primera, si se quiere ejecutar de nuevo hay que liberar los recursos con el boton repeat del jupyter
#posicionado dos cuadros a la derecha del "run"

from datetime import datetime
print("Proceso de: "+str(nums_reg)+" x "+str(maxC)+" columnas")
now=datetime.now()
print ("hora de inicio (apriori): "+str(now.time()))
association_rules = apriori(records, min_support=0.0053, min_confidence=0.2, min_lift=3, min_length=2)
now=datetime.now()
print ("hora de inicio (list): "+str(now.time()))
association_results = list(association_rules)
now=datetime.now()
print ("hora finalización: "+str(now.time()))
```
